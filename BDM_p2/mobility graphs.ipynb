```python
import pandas as pd
import seaborn as sns
import sqlite3
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Global_Mobility_Report.csv', parse_dates=["date"])
dx=df#backup dataset
dx.columns
conn = sqlite3.connect("global_mobility.db")
df.to_sql('gm_table', conn)
```

```python
fig, ax1 = plt.subplots()
plt.xticks(rotation=90) 
color = 'tab:red'
ax1.set_xlabel('Number of days')
ax1.set_ylabel('retail_and_recreation_percent_change_from_baseline', color=color)
ax1.plot(df3.index, df3['retail_and_recreation_percent_change_from_baseline'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('New Cases', color=color)  # we already handled the x-label with ax1
ax2.plot(df3.index, df3['new_cases'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
